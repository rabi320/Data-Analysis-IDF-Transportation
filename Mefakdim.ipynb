# שאלון מפקדים

```python
import pandas as pd
#all
all_data = pd.read_excel("Data2/תכנון לו''ז  תחקור.xlsx", "Sheet1")
all_data.head()
```

# ממוצע ע"פ דרג

```python
df = all_data.groupby(by = all_data.columns[3]).agg('mean')
df.loc["כולל",:] = all_data[df.columns].mean(axis = 0)
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
plt.style.use('ggplot')

plot1 = df.T
colors = cm.Set1(np.linspace(0,1,len(plot1)))
plot1.index = [i[::-1] for i in plot1.index]
plot1.columns = [i[::-1] for i in plot1.columns]
title = 'דירוג ממוצע ע"פ סוג השאלה: דרג'
#plot1.index
plot1.plot(kind = "bar", color = colors, figsize = (20,6))

plt.xticks( size = 12)
plt.yticks(size = 12)

plt.title(title[::-1], size = 20)
plt.show()
```

# ממוצע ע"פ שאט"ל

```python
df = all_data.groupby(by = all_data.columns[0]).agg('mean')
df.loc["כולל",:] = all_data[df.columns].mean(axis = 0)
df
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
plt.style.use('ggplot')

plot1 = df.T
colors = cm.gist_rainbow(np.linspace(0,1,len(plot1)))
plot1.index = [i[::-1] for i in plot1.index]
plot1.columns = [i[::-1] for i in plot1.columns]
title = 'דירוג ממוצע ע"פ סוג השאלה: שאט"ל'
#plot1.index
plot1.plot(kind = "bar", color = colors, figsize = (20,6))

plt.xticks( size = 12)
plt.yticks(size = 12)

plt.title(title[::-1], size = 20)
plt.show()
```
