# שאלון מפקדים

```python
import pandas as pd
#all
all_data = pd.read_excel("Data2/תכנון לו''ז  תחקור.xlsx", "Sheet1")
all_data
```

# ממוצע ע"פ דרג

```python
df = all_data.groupby(by = all_data.columns[3]).agg('mean')
df.loc["כולל",:] = df.mean(axis = 0)
df
```

# ממוצע ע"פ בתים

```python
df = all_data.groupby(by = all_data.columns[1]).agg('mean')
df.loc["כולל",:] = df.mean(axis = 0)
df
```

# ממוצע ע"פ שאט"ל

```python
df = all_data.groupby(by = all_data.columns[0]).agg('mean')
df.loc["כולל",:] = df.mean(axis = 0)
df
```
