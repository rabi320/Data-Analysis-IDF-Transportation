# שאלון מפקדים

```python
import pandas as pd
#all
all_data = pd.read_excel("Data2/תכנון לו''ז  תחקור.xlsx", "Sheet1")
all_data.head()
```

# תצוגה כללית מי משתמש ולא משתמש בשאטל

```python
import matplotlib.pyplot as plt
import seaborn as sns
#colors
my_colors = ["orange","yellow","red","lime","pink",
         "mediumpurple","skyblue","cornflowerblue","gray","gainsboro",
         "gold","turquoise","chocolate","forestgreen","orchid"]

#define data
df =all_data[all_data.columns[0]].value_counts()
labels = df.index
sizes = df.values

labels= [label[::-1] for label in labels]
label_dict = dict(zip(labels,df.values/sum(df.values)))

colors = dict(zip(labels,my_colors[10:13]))

#create pie chart
plt.figure(figsize  = (15,15))
plt.pie(sizes, labels = [label if label_dict[label]>0.015  else "" for label in labels],colors = [colors[p] for p in labels],autopct=lambda p: '{:.2f}%'.format(round(p,2)) if p > 3 else '',textprops={'fontsize': 25})
plt.title(all_data.columns[0][::-1], size = 30)
plt.legend(labels,fontsize = 14)
plt.show()
```

# תצוגה לפי בתים מי משתמש ולא משתמש בשאטל

```python
batim = all_data[all_data.columns[1]].unique().tolist()
for i,j in enumerate(batim):
    df =all_data[all_data[all_data.columns[1]]==batim[i]][all_data.columns[0]].value_counts()
    labels = df.index
    sizes = df.values

    labels= [label[::-1] for label in labels]
    label_dict = dict(zip(labels,df.values/sum(df.values)))

    colors = dict(zip(labels,my_colors[10:13]))

    #create pie chart
    plt.figure(figsize  = (15,15))
    plt.pie(sizes, labels = [label if label_dict[label]>0.015  else "" for label in labels],colors = [colors[p] for p in labels],autopct=lambda p: '{:.2f}%'.format(round(p,2)) if p > 3 else '',textprops={'fontsize': 25})
    plt.title(batim[i][::-1], size = 30)
    plt.legend(labels,fontsize = 14)
    plt.show()
```

# השוואה בין הבא"חים שמשתמשים באפליקציה

```python
from matplotlib import cm
import numpy as np
plt.style.use('ggplot')
df =all_data.groupby([all_data.columns[1],all_data.columns[0]]).agg('mean')
df.loc[("כולל","כולל"),:] = all_data[df.columns].mean(axis = 0)
df = df.drop([('בית הבה"דים', 'כן'),
 ('בית הבה"דים', 'לא'),
 ('בית היחידה המבצעית', 'כן'),
 ('בית היחידה המבצעית', 'לא'),
 ('בית היחידה המבצעית', 'לא ידוע'),
 ('בית ח"א', 'כן'),
 ('בית ח"א', 'לא'),('קריית ההדרכה', 'כן')],axis = 0)

plot1 = df.T
colors = cm.Set1(np.linspace(0,1,len(plot1)))
plot1.index = [i[::-1] for i in plot1.index]
plot1.columns = [i[::-1] for i in plot1.columns]
title = 'דירוג ממוצע ע"פ בא"חים שמשתמשים ולא משתמשים בשאט"ל'
#plot1.index
plot1.plot(kind = "bar", color = colors, figsize = (20,6))

plt.xticks( size = 16)
plt.yticks(size = 16)
legend_list = ['בא"חים שמשתמשים בשאטל','בא"חים שלא משתמשים בשאטל','ממוצע כולל']
plt.legend([a[::-1] for a in legend_list],fontsize = 18)
plt.title(title[::-1], size = 20)
plt.show()
```
